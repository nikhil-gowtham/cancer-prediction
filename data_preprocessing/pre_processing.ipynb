## Read Dataset

```python
# import required libraries
import pickle

import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
# read dataset to dataframe
df= pd.read_csv(r"C:\Users\nikhi\Personal\bits\3\mlops\assignment1\cancer-prediction\data\Cancer_Data.csv")
```

```python
df.head()
```

```python
df.info()
```

## Remove unnecessary columns and rows

- We do not need 'id' for our prediction. Hence, we can remove it.
- 'diagnosis' is the labels and not features.
- 'Unnamed: 32' is also not needed as it has only Nan values.

```python
data = df.drop(['id', 'diagnosis','Unnamed: 32'], axis=1)
```

```python
data.head()
```

Remove records which have null values in it (if any).

```python
data = data.dropna()
```

## Label Encoder

The labels are "M" (Malignant - Benign) or "B" (Benign - Malignant). We need to encode them for our model training.

```python
labels = df['diagnosis']
```

```python
# create a LabelEncoder object
label_encoder = LabelEncoder()

# fit and transform the data
encoded_labels = label_encoder.fit_transform(labels)
```

```python
encoded_labels
```

```python
# save label encoder
with open('label_encoder.pkl', 'wb') as f:
    pickle.dump(label_encoder, f)
```

## Standard Scaler

```python
# create a StandardScaler object
scaler = StandardScaler()

# fit the scaler to the data and transform it
scaled_data = scaler.fit_transform(data)
```

```python
# save scaler
with open('standard_scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
```

## Save Preprocessed Data

```python
# create dataframe with final data
final_df = pd.DataFrame(scaled_data, columns=data.columns)
final_df['diagnosis'] = encoded_labels
```

```python
final_df.head()
```

```python
# save the scaled data to a new CSV file
final_df.to_csv("scaled_data.csv", index=False)
```
